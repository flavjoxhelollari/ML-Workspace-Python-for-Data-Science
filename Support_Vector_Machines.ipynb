# Notes - Support Vector Machines

- 1980's marked an important era on improvements on learning algorithms, as new computational theories evolved. 
- Non linear separable regions were finally separated by kernel functions, and neural nets avoided the local optima problem by using quadratic functions.
- Structure of SVM : optimal hyperplane for linear separable data; use kernel transformation otherwise.
- def: Support Vector - data points that lie closest to the decision surface; most difficult to classify; have explicit impact on the optimum position of decision surface. Note that SVM finds an optimal hyperplane to separate the data, as there are multiple possible solutions to the classification problem. 
- SVM can maximize the margin around the hyperplane. Support vectors, subset of training samples, are in full control of the decision function. 
- The problem of finding the optimal hyper plane is a constrained
optimization problem and can be solved by optimization
techniques (use Lagrangian multipliers). Can be viewed as quadratic programming problem, so the surface is a paraboloid with a global optima. 
- image.png

- Relying on Karush-Kuhn-Tucker Theorem, one can convert the above (primal) problem, into an equivalent dual problem. This way, we can find the solution to the constrained optimization problem, by simply calculating the inner product of the training point and a test point, say x_i, x_j. 
)

- In a geometrical POV, consider support vectors, belonging to the two different classes.  Consider the first case: 2 very simmilar vectors x_i, x_j, of different classes, maximize the margin width; 2 vectors of the same class, similar to each other are redundant; 2 orthogonal vectors are practically useless.
- image.png
- In other words, map x-x**2, in order to gain linear separability.
- Kernel trick: the kernel function defines nner products in the transformed space, i.e. defines similarity.
- Some kernel tricks can include polynomial, radial basis, or sigmoid transformations. Also, note that kernels generalize the idea of 'inner product similarity'. 
- Soft Margin: In SV Classifier, we are allowed to have some points to violate the margin, but not too many. 
- A convention, is to note the width of the Maxim. Margin Classifier as . For SVC,  ~ 1/K. This is how one is able to tune the bias-variance tradeoff. K ~ Bias, K ~ 1/Variance. Note that in sklearn, K = 1/C.
- Loss function used in this perspective is called hinge loss. The idea is to minimize w such that the sum of (max[0, 1 dist. from H_i] + width of the margin) is maximized. Compared to Logistic Regeression Loss, the loss in Hinge, of the points on the correct side of the margin is zero.

Case 1 - SVM with linear kernel has more bias, but SVM with RBF kernel has more variance.

```python
from tqdm import tqdm

import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import ipywidgets as widgets
from ipywidgets import interact, fixed
from mpl_toolkits import mplot3d

# use seaborn plotting defaults
import seaborn as sns; sns.set_style('white')
from sklearn.datasets import make_blobs, make_circles
from sklearn.svm import SVC # "Support vector classifier"
```

```python
X, y = make_blobs(n_samples=50, centers=2,
                  random_state=0, cluster_std=0.60)
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='autumn');
```

```python
model = SVC(kernel='linear', C=1E10)
model.fit(X, y)
```

```python
print(model.support_vectors_)
print(model.C)
print(model.decision_function_shape)
print(model.gamma)
print(model.kernel)
```
