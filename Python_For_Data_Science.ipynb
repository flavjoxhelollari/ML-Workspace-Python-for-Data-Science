Numpy Arrays: 
1. Attributes : Determining the size*(np.size)*, shape *(np.shape),* memory consumption, and data types of arrays. *(np.ndim)*

2. Indexing of arrays
Getting and setting the value of individual array elements

3. Slicing of arrays
Getting and setting smaller subarrays within a larger array
    `*x[start:stop:step]*` 
  Multidimensional slices work in the same way, with multiple slices separated by commas. To access single row or column: 

```
print(x2[:, 0])
 first column of x2
print(x2[0, :])
 first row of x2
```
Important note: Array slicing returns views of the original array, and not copies. This means changes are reflected in the original whole array.  

4. Reshaping of arrays
Changing the shape of a given array.
The most flexible way of doing this is with the **reshape()** method.



5. Joining and splitting of arrays
Combining multiple arrays into one, and splitting one array into many. Concatenation of arrays is made possible via the following methods : 

```
np.concatenate , np.vstack , and np.hstack 
```
They have an optional axis as parameter. 
For example: 


```
x = np.array([1, 2, 3])
grid = np.array([[9, 8, 7],[6, 5, 4]])
 vertically stack the arrays
np.vstack([x, grid])
# output is the array : [[1,2,3],
                          [9,8,7],
                          [6,5,4]]

# horizontally stack the arrays
y = np.array([[99],[99]])
np.hstack([grid, y])
# output is the array : [[9,8,7,99]
                         [6,5,4,99]]

```

Splitting is done via:


```
np.split , np.hsplit , and np.vsplit
```

# Pandas


````python
import pandas as pd
````

In a sense, pandas objects can be thought of as enhanced numpy structures. Pandas provides a host of useful tools, methods, and functionality on top of the basic
data structures.

1. Pandas Series: a one-dimensional array of indexed data. It can be reated from a list or array as follows:data= pd.Series([1,2,3,4,5,6]).
Accessed similarly by `[]`.

There is an essential difference between numpy lists and series in pandas. The later, can be thought of as specialized dictionaries in python, since we can specify indexes for the list, and it can be of any data type.


2. Pandas DataFrame is teh next fundamental structure in pandas. DataFrame
is an analog of a two-dimensional array with both flexible row indices and flexible column names. Relevant attributes : 

```
dataframe.index or dataframe.columns 
```
we can also think of a DataFrame as a specialization of a dictionary. Where
a dictionary maps a key to a value, a DataFrame maps a column name to a Series of
column data. How to construct a pandas dataframe? We can do it from a single Series object, from a list of dicts, from a d numpy array, etc...
Both, Series and DataFrame have an **index** object. It can be thought of either as an **immutable** array or as an ordered set. 


Pandas also provides special indexers, which provide slicing interface to the data. First, the loc attribute allows indexing and slicing that always references the explicit index. The iloc attribute allows indexing and slicing that always references the implicit
Python-style index. A third indexing attribute, ix , is a hybrid of the two, and for Series objects is equiva
lent to standard [] -based indexing.

**Operating on Data**
Any NumPy ufunc works on pandas Series and DataFrame.


**Handling Missing Data**
Pandas uses *sentinels* for missing data; NaN and None. 

None can only be used in arrays with data types 'object'. 

NaN stands for Not a Number. It is a special floating-point value. Operation on every object with a NaN one, results into NaN.

String data is alwyas stored in an object dtype.

Tere are several useful
methods for detecting, removing, and replacing null values in Pandas data structures.


```
isnull() - Generate a Boolean mask indicating missing values
notnull() - Opposite of isnull()
dropna() - Return a filtered version of the data
fillna() - Return a copy of the data with missing values filled or imputed
```

Note that, By default, dropna() will drop all rows in which any null value is present. You can specify axis = 1 to drop the columns with NA values. 

On filling the NA values, you can do the followings: `data.fillna(0)`;
you can specify a forward-fill to propagate the previous value forward. 

```
data.fillna(method='ffill') for forward fill and
data.fillna(method='bfill')for backward fill
```
For DataFrames, it is similar, but we can also specify the axis.
